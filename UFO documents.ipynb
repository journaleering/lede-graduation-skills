## Wrangling your data

### Make a list of all `.tif` filenames  `2 points`

```python
import glob

os.chdir('CIAUFOCD-FULL-UNTOUCHED')
filenames = glob.glob('*.tif')
os.chdir('..')

filenames
```

### Create a dataframe of all documents

One column should be the filename, the other should be all of the OCR'd content. **Tika isn't the fastest, so this will probably take a while.**

```python
os.chdir('CIAUFOCD-FULL-UNTOUCHED')
contents = [pytesseract.image_to_string(filename) for filename in filenames]
os.chdir('..')
```

```python
df1 = pd.DataFrame({
    'filename': filenames,
    'content': contents
})
df1.content = df1.content.replace('\n',' ', regex=True)
df1
```
